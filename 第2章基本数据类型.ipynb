### 2.2.5　变量的存储

id()函数可以显示一个变量在内存中的地址

```python
a = 123
print(id(a))
a = 456
print(id(a))
```

同一个地址空间可以有两个变量名，甚至多个变量名

```python
a = 123
print(id(a))
b = 123
print(id(b))
a = 456
print(id(a))
b = 456
print(id(b))
```

* 如何实现两个变量值的互换?

```python
a = 123
b = 456
c = a
a = b
b = c
print(a)
print(b)
```

```python
a = 123
b = 456
a,b = b,a
print(a)
print(b)
```
